```
import numpy as np
import matplotlib.pyplot as plt

def u(t):
    return np.where(t >= 0, 1, 0)

t = np.linspace(-10, 10, 1000)
x1 = u(t) + 2*u(t) - u(t - 2) - 2*u(-t + 1)

plt.plot(t, x1)
plt.title('x(t) = u(t) + 2u(t) - u(t-2) - 2u(-t+1)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid(True)
plt.show()
```

```
x2 = np.exp(-t**2) * np.cos(2 * np.pi * t)

plt.plot(t, x2)
plt.title('x(t) = e^{-t^2} cos(2Ï€t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid(True)
plt.show()
```

```
x3 = np.exp(-t) * u(t)

plt.plot(t, x3)
plt.title('x(t) = e^{-t} u(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid(True)
plt.show()
```

```
x4 = np.exp(-t) * u(t) + np.exp(5*t) * u(-t)

plt.plot(t, x4)
plt.title('x(t) = e^{-t} u(t) + e^{5t} u(-t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid(True)
plt.show()
```

```
x5 = np.cos(0.75 * np.pi * t) + 0.5 * np.sin(0.75 * np.pi * t)

plt.plot(t, x5)
plt.title('x(t) = cos(0.75Ï€t) + 0.5 sin(0.75Ï€t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid(True)
plt.show()
```

```
x6 = np.exp(-t) * np.sin(8 * np.pi * t + 0.23 * np.pi) * u(t)

plt.plot(t, x6)
plt.title('x(t) = e^{-t} sin(8Ï€t + 0.23Ï€) u(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid(True)
plt.show()
```

```
x7 = np.exp(-t) * u(t) + np.exp(-4*t) * u(t)

plt.plot(t, x7)
plt.title('x(t) = e^{-t} u(t) + e^{-4t} u(t)')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.grid(True)
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-10, 10, 1000)

alphas = [-10, -1, -0.2, 0, 0.5, 1, 10]

plt.figure(figsize=(10, 6))
for alpha in alphas:
    x_alpha = np.exp(alpha * t)
    plt.plot(t, x_alpha, label=f'Î± = {alpha}')

plt.title('Exponential function e^{Î±t} for different Î± values')
plt.xlabel('t')
plt.ylabel('e^{Î±t}')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)  # x-axis
plt.axvline(0, color='black',linewidth=0.5)  # y-axis
plt.legend()
plt.show()
```
